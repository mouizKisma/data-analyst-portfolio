# Exporting X_train1 and y_train1

```python
# Rename columns for df1 dataframes
df1_bearing1_ch1.rename(columns={'bearing1_ch1': 'value'}, inplace=True)
df1_bearing1_ch2.rename(columns={'bearing1_ch2': 'value'}, inplace=True)
df1_bearing2_ch3.rename(columns={'bearing2_ch3': 'value'}, inplace=True)
df1_bearing2_ch4.rename(columns={'bearing2_ch4': 'value'}, inplace=True)
df1_bearing3_ch5.rename(columns={'bearing3_ch5': 'value'}, inplace=True)
df1_bearing3_ch6.rename(columns={'bearing3_ch6': 'value'}, inplace=True)
df1_bearing4_ch7.rename(columns={'bearing4_ch7': 'value'}, inplace=True)
df1_bearing4_ch8.rename(columns={'bearing4_ch8': 'value'}, inplace=True)
# List of all dataframes with their names
dataframes = [
    ('df1_bearing1_ch1', df1_bearing1_ch1),
    ('df1_bearing1_ch2', df1_bearing1_ch2),
    ('df1_bearing2_ch3', df1_bearing2_ch3),
    ('df1_bearing2_ch4', df1_bearing2_ch4),
    ('df1_bearing3_ch5', df1_bearing3_ch5),
    ('df1_bearing3_ch6', df1_bearing3_ch6),
    ('df1_bearing4_ch7', df1_bearing4_ch7),
    ('df1_bearing4_ch8', df1_bearing4_ch8)
]
```

```python
import numpy as np
import pandas as pd

def create_windows(df, window_size=2048, overlap=0, 
                   value_col='value', label_col='label', 
                   group_col='timestamp'):
    
    X, y = [], []
    step = int(window_size * (1 - overlap))
    
    for group_key, group in df.groupby(group_col):
        values = group[value_col].values
        label = group[label_col].iloc[0]
        
        for start in range(0, len(values) - window_size + 1, step):
            window = values[start:start+window_size]
            X.append(window.reshape(-1, 1))
            y.append(label)
    
    return np.array(X), np.array(y)


# Process all dataframes
X_list1, y_list1 = [], []

for name, df in dataframes:
    print(f"Processing {name}... ", end='')
    X, y = create_windows(df, window_size=2048, overlap=0, value_col='value')
    X_list1.append(X)
    y_list1.append(y)
    print(f"Created {X.shape[0]} windows")

# Combine all windows
X_train1 = np.concatenate(X_list1, axis=0)
y_train1 = np.concatenate(y_list1, axis=0)

print(f"\nFinal combined shape: X={X_train1.shape}, y={y_train1.shape}")
print(f"Total windows: {len(X_train1)}")
print(f"Label distribution: {np.unique(y_train1, return_counts=True)}")
```

```python
np.save('X_train1_no_overlap.npy', X_train1)
np.save('y_train1_no_overlap.npy', y_train1)
```
